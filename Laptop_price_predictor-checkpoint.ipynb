## Data cleaning of unuseful columns

```python
# data cleaning of unuseful columns
df1.drop(['brand','name','img_link','display(in inch)','no_of_ratings','no_of_reviews'], axis=1, inplace=True)
df1.head()
```

```python
df1.head()
```

## Label Encoding

```python
categorical_features =df1.columns[(df1.dtypes == 'object') == True].to_list()
print(categorical_features)
```

```python
for feature in categorical_features:
    uniq = df1[feature].unique()
    
    new_feature = []
    for el in df1[feature]:
        new_feature.append(len(uniq) - np.where(uniq==el)[0][0])
    
    df1[feature] = new_feature
    
df1.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(df1.drop(['price(in Rs.)'], axis=1), df1['price(in Rs.)'], test_size=0.3, random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score

# Initialize the model
model = RandomForestRegressor()

# Train the model
model.fit(X_train, Y_train)

# Make predictions
predictions = model.predict(X_test)

# Evaluate the model
r2 = r2_score(Y_test, predictions)
print("R² score:", r2)
```

```python
plt.scatter(np.linspace(0, len(predictions), len(predictions)), predictions, alpha=0.5, label='Predictions')
plt.scatter(np.linspace(0, len(Y_test), len(Y_test)), Y_test, alpha=0.5, label='True')
plt.legend()
```
